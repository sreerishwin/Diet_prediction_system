```python
import tkinter as tk
from tkinter import ttk
from tkinter import *
import tkinter.font as tkFont

import sys
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.ensemble import RandomForestClassifier

class MainWindow:
    def __init__(self, root):
        root.configure(bg="#071e26")
        root.title("Diet Recommender")
        width=900
        height=410
        screenwidth = root.winfo_screenwidth()
        screenheight = root.winfo_screenheight()
        alignstr = '%dx%d+%d+%d' % (width, height, (screenwidth - width) / 2, (screenheight - height) / 2)
        root.geometry(alignstr)
        root.resizable(width=False, height=False)

        self.meal=tk.IntVar()
        self.age, self.height, self.weight= tk.IntVar(), tk.DoubleVar(), tk.DoubleVar()
        self.goal, self.sex, self.activity=tk.IntVar(), tk.IntVar(), tk.DoubleVar()

        #Main Heading
        ft = tkFont.Font(family='Times',size=22)
        self.label1=tk.Label(root, font=ft,justify = LEFT, text = "PERSONALISED DIET RECOMMENDATION SYSTEM",fg="white",bg="#071e26")
        self.label1.place(x=135,y=10)

        #choose Sex
        ft = tkFont.Font(family='Times',size=10)
        self.label2=tk.Label(root,font = ft, justify = LEFT, text = "Sex : ",fg="white",bg="#071e26")
        self.label2.place(x=30,y=60)
        self.rbmale = Radiobutton(root, text='Male', variable=self.sex, value=0, cursor="hand2",fg="white",bg="#071e26")
        self.rbmale.place(x=70, y=60)
        self.rbfemale = Radiobutton(root, text='Female', variable=self.sex, value=1, cursor="hand2",fg="white",bg="#071e26")
        self.rbfemale.place(x=120, y=60)

        #Get Age
        ft = tkFont.Font(family='Times',size=10)
        self.label3=tk.Label(root,font = ft, justify = LEFT, text = "Age : ",fg="white",bg="#071e26")
        self.label3.place(x=30,y=90)

        self.entryage=tk.Entry(root, font = ft, justify = LEFT, textvariable = self.age,bg="#2596be",fg="white")
        self.entryage.place(x=80,y=90,width=200,height=30)

        #Get Height
        ft = tkFont.Font(family='Times',size=10)
        self.label4=tk.Label(root,font = ft, justify = LEFT, text = "Height : ",fg="white",bg="#071e26")
        self.label4.place(x=30,y=120)

        self.entryheight=tk.Entry(root, font = ft, justify = LEFT, text=" ",textvariable = self.height,bg="#2596be",fg="white")
        self.entryheight.place(x=80,y=120,width=200,height=30)

        #Get Weight
        ft = tkFont.Font(family='Times',size=10)
        self.label5=tk.Label(root,font = ft, justify = LEFT, text = "Weight : ",fg="white",bg="#071e26")
        self.label5.place(x=30,y=150)

        self.entryweight=tk.Entry(root, font = ft, justify = LEFT, textvariable = self.weight,bg="#2596be",fg="white")
        self.entryweight.place(x=80,y=150,width=200,height=30)

        #choose Goal
        ft = tkFont.Font(family='Times',size=10)
        self.label6=tk.Label(root,font = ft, justify = LEFT, text = "Goal : ",fg="white",bg="#071e26")
        self.label6.place(x=30,y=190)
        self.rbloss = Radiobutton(root, text='Weight Loss', variable=self.goal, value=0, cursor="hand2",fg="white",bg="#071e26")
        self.rbloss.place(x=90, y=190)
        self.rbgain = Radiobutton(root, text='Weight Gain', variable=self.goal, value=1, cursor="hand2",fg="white",bg="#071e26")
        self.rbgain.place(x=190, y=190)
        self.rbhealthy = Radiobutton(root, text='Healthy', variable=self.goal, value=2, cursor="hand2",fg="white",bg="#071e26")
        self.rbhealthy.place(x=290, y=190)

        #choose Activity
        ft = tkFont.Font(family='Times',size=8)
        self.label7=tk.Label(root,font=ft, justify=LEFT, text = "Activity : ",fg="white",bg="#071e26")
        self.label7.place(x=400,y=80)
        self.rbbreakfast = Radiobutton(root, font= ft,justify=LEFT,text='Sedentary(little to no exercise)', variable=self.activity, value=1.2, cursor="hand2",fg="white",bg="#071e26")
        self.rbbreakfast.place(x=400, y=105)
        self.rblunch = Radiobutton(root, font=ft,justify=LEFT,text='Lightly active\n(light exercise/sports 1-3 times a week)', variable=self.activity, value=1.375, cursor="hand2",fg="white",bg="#071e26")
        self.rblunch.place(x=400, y=140)
        self.rbdinner = Radiobutton(root, font=ft,justify=LEFT,text='Moderately active\n(moderate exercise/sports 3-5 times a week)', variable=self.activity, value=1.55, cursor="hand2",fg="white",bg="#071e26")
        self.rbdinner.place(x=400, y=175)
        self.rbdinner = Radiobutton(root, font=ft,justify=LEFT,text='Very active\n(hard exercise/sports 6-7 times a week)', variable=self.activity, value=1.725, cursor="hand2",fg="white",bg="#071e26")
        self.rbdinner.place(x=400, y=210)
        self.rbdinner = Radiobutton(root, font=ft,justify=LEFT,text='Extra active\n(very hard exercise/sports\n&physical job or 2x training)', variable=self.activity, value=1.9, cursor="hand2",fg="white",bg="#071e26")
        self.rbdinner.place(x=400, y=245)

        #choose Meal
        ft = tkFont.Font(family='Times',size=10)
        self.label7=tk.Label(root,font = ft, justify = LEFT, text = "Meal : ",fg="white",bg="#071e26")
        self.label7.place(x=30,y=220)
        self.rbbreakfast = Radiobutton(root, text='Breakfast', variable=self.meal, value=1, cursor="hand2",fg="white",bg="#071e26")
        self.rbbreakfast.place(x=90, y=220)
        self.rblunch = Radiobutton(root, text='Lunch', variable=self.meal, value=2, cursor="hand2",fg="white",bg="#071e26")
        self.rblunch.place(x=190, y=220)
        self.rbdinner = Radiobutton(root, text='Dinner', variable=self.meal, value=3, cursor="hand2",fg="white",bg="#071e26")
        self.rbdinner.place(x=290, y=220)

        #Get Diet Button
        #style = ttk.Style()
        #style.configure("Custom.TButton", relief="flat", borderwidth=0, bordercolor="gray", padding=10, focuscolor="blue")
        self.buttontel=tk.Button(root,font = ft, text = "Get Diet",fg="white",bg="#165a72",)
        self.buttontel.place(x=30,y=270,width=110,height=30)
        self.buttontel["command"] = self.calculate

        #Quit
        self.buttontel=tk.Button(root,font = ft, text = "Quit",fg="white",bg="#165a72")
        self.buttontel.place(x=160,y=270,width=110,height=30)
        self.buttontel["command"] = self.quit

    def calculate(self):
        window=Tk()
        window.geometry('1250x550')
        window.title("Diet")
        window.configure(bg ="white")

        tree = ttk.Treeview(window, columns=tuple(range(0, 12)), show="headings")
        scrollbar = ttk.Scrollbar(window, orient="vertical", command=tree.yview)
        scrollbar.pack(side="right", fill="y")
        tree.configure(yscrollcommand=scrollbar.set)

        tree["columns"] = tuple(range(0, 12))

        age, weight, height = self.age.get(), self.weight.get(), self.height.get()
        sex, goal, meal = self.sex.get(), self.goal.get(), self.meal.get()

        bmi = (weight/(height**2)*10)
        #BMR according to Harris-Benedict Formula 
        bmrmale=66.47+(13.75*weight)+(5.003*height*100)-(6.755*age)
        bmrfemale=655.1+(9.563*weight)+(1.850*height*100)-(4.675*age)

        bmr = bmrmale if sex==0 else bmrfemale
        #Calculating Daily calorie requirement
        amr = bmr*self.activity.get()

        bmiprompt=''
        if ( bmi < 16):
            bmiprompt = "Severely Underweight"
            clbmi=4
        elif ( bmi >= 16 and bmi < 18.5):
            bmiprompt = "Underweight"
            clbmi=3
        elif ( bmi >= 18.5 and bmi < 25):
            bmiprompt = "Healthy"
            clbmi=2
        elif ( bmi >= 25 and bmi < 30):
            bmiprompt = "Overweight"
            clbmi=1
        elif ( bmi >=30):
            bmiprompt = "Severely Overweight"
            clbmi=0

        basic= f"Age: {age} Years\nWeight: {weight} Kg\nHeight: {height} m\n"
        basic+= f"BMI: {bmi}\nBMI remark:{bmiprompt}\nBasal Matabolic Rate: {bmr} Calories\nDaily Calorie Requirement: {amr} Calories\n\nDiet :"

        Label(window,font = tkFont.Font(family='Times',size=10), bg='white', text=basic, justify = LEFT).place(x=30, y=20)

        data=pd.read_csv('input.csv')

        if meal==1:Mealdata=data['Breakfast'].to_numpy()
        elif meal==2:Mealdata=data['Lunch'].to_numpy()
        else:Mealdata=data['Dinner'].to_numpy()
        
        Food_itemsdata=data['Food_items']
        Mealseparated, MealseparatedID = [], []

        tree.heading(0, text="Food")
        tree.column(0, width=100)
        for i in range(1, 12):
            tree.heading(i, text=data.columns[i+4 ])
            tree.column(i, width=100)
        
        for i in range(len(Mealdata)):
            if Mealdata[i]==1:
                Mealseparated.append(Food_itemsdata[i])
                MealseparatedID.append(i)
        
        MealseparatedIDdata = data.iloc[MealseparatedID].T
        val=list(np.arange(5,16))
        Valapnd=[0]+val
        MealseparatedIDdata=MealseparatedIDdata.iloc[Valapnd]
        MealseparatedIDdata=MealseparatedIDdata.T.to_numpy()  #Just nutritional Value table

        #Find age Group
        for lp in range (0,80,20):
            for i in np.arange(lp,lp+20): 
                if(i == age):
                    agecl=round(lp/20)    
        ti=(clbmi+agecl)/2

        ## K-Means Based  Dinner Food
        Datacalorie=MealseparatedIDdata[:,1:len(MealseparatedIDdata)]
        X = np.array(Datacalorie)
        kmeans = KMeans(n_init=10, n_clusters=3, random_state=0).fit(X)
        meallabels=kmeans.labels_

        #dataTog=pd.read_csv('inputfin.csv').T
        dataTog=pd.DataFrame(Datacalorie.T)

        bmicls=[0,1,2,3,4]
        agecls=[0,1,2,3,4]

        dcat=[]
        if goal==0:
            dcat=dataTog.iloc[[1,2,7,8]].T.to_numpy()
            dcat=dcat[1:,0:len(dcat)]
            dfin=np.zeros((len(dcat)*5,6),dtype=np.float32)
            X_test=np.zeros((len(dcat),6),dtype=np.float32)
        elif goal==1:
            dcat= dataTog.iloc[[0,1,2,3,4,7,9,10]].T.to_numpy()
            dcat=dcat[1:,0:len(dcat)]
            dfin=np.zeros((len(dcat)*5,10),dtype=np.float32)
            X_test=np.zeros((len(dcat),10),dtype=np.float32)
        elif goal==2:
            dcat = dataTog.iloc[[1,2,3,4,6,7,9]].T.to_numpy()
            dcat=dcat[1:,0:len(dcat)]
            dfin=np.zeros((len(dcat)*5,9),dtype=np.float32)
            X_test=np.zeros((len(dcat),9),dtype=np.float32)

        t,yt=0, []

        for zz in range(5):
            for jj in range(len(dcat)):
                valloc=list(dcat[jj])
                valloc.append(bmicls[zz])
                valloc.append(agecls[zz])
                dfin[t]=np.array(valloc)
                yt.append(meallabels[jj])
                t+=1

        for jj in range(len(dcat)):
            valloc=list(dcat[jj])
            valloc.append(agecl)
            valloc.append(clbmi)
            X_test[jj]=np.array(valloc)*ti

        X_train= dfin
        y_train=yt
                    
        clf=RandomForestClassifier(n_estimators=100)
        clf.fit(X_train,y_train)
        y_pred=clf.predict(X_test)
        
        for ii in range(len(y_pred)):
            if y_pred[ii]==2:
                food=data.loc[data["Food_items"] == Food_itemsdata[ii]].to_numpy()[0]
                tree.insert(parent='', index=tk.END, values=(Food_itemsdata[ii], food[5], food[6], food[7], food[8], food[9], food[10], food[11], food[12], food[13], food[14], food[15]))
                pass
        
        tree.place(x=30, y=250)
        
    def quit(self):
        sys.exit(0)


def main():
    root = tk.Tk()
    gui = MainWindow(root)  
    root.mainloop()

if __name__ == '__main__':
    main()
```
